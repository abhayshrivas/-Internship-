```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

url = 'https://raw.githubusercontent.com/dsrscientist/dataset4/main/Grades.csv'
data = pd.read_csv('https://raw.githubusercontent.com/dsrscientist/dataset4/main/Grades.csv')

columns = ['CGPA', 'PH-121', 'HS-101']
x = data[columns]

for column in columns:
    x[column] = pd.to_numeric(x[column], errors='coerce')
    
fig = plt.figure(figsize=(20,10))
plotnumber = 1

for column in columns:
    if plotnumber <= 4:
        ax = plt.subplot(2, 2, plotnumber)
        sns.distplot(x[column].dropna())
        plt.xlabel(column, fontsize=20)
        plt.ylabel('CGPA')
    plotnumber += 1

plt.tight_layout()
plt.show()
```
