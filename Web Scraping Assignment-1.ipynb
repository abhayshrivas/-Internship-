```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://en.wikipedia.org/wiki/Main_Page')
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
headers_list = soup.find_all("h1")+soup.find_all("h2")+soup.find_all("h3")+soup.find_all("h4")+soup.find_all("h5")+soup.find_all("h6")
```

```python
for tag in headers_list:
    print(tag.name,tag.text)
```
